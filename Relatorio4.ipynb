```python
import pandas as pd
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt
```

```python
caract_channel1_inf_plato = pd.read_csv("caract_channel1_inf_plato.txt", sep=" ")
caract_channel2_sup_plato = pd.read_csv("caract_channel2_sup_plato.txt", sep=" ")
```

```python
caract_channel1_inf_plato
```

```python
caract_channel2_sup_plato
```

```python
caract_channel1_inf_plato['Cmean'] = (caract_channel1_inf_plato['C1']+caract_channel1_inf_plato['C2']+caract_channel1_inf_plato['C3'])/3.
```

```python
caract_channel2_sup_plato['Cmean'] = (caract_channel2_sup_plato['C1']+caract_channel2_sup_plato['C2']+caract_channel2_sup_plato['C3']+caract_channel2_sup_plato['C4']+caract_channel2_sup_plato['C5']+caract_channel2_sup_plato['C6'])/6.
```

```python
caract_channel1_inf_plato
```

```python
caract_channel2_sup_plato
```

```python
caract_channel2_sup_plato['Cerror'] = (caract_channel2_sup_plato['Cmean'])**0.5/6.
```

```python
caract_channel1_inf_plato['Cerror'] = (caract_channel1_inf_plato['Cmean'])**0.5/3.
```

```python
caract_channel2_sup_plato
```

```python
caract_channel1_inf_plato
```

```python
df = caract_channel2_sup_plato
```

```python
limiar_10_ch2 = pd.concat((df['HV'].loc[(df["Limiar"]==10)],df['Cmean'].loc[(df["Limiar"]==10)],df['Cerror'].loc[(df["Limiar"]==10)]),axis=1)
limiar_20_ch2 = pd.concat((df['HV'].loc[(df["Limiar"]==20)],df['Cmean'].loc[(df["Limiar"]==20)],df['Cerror'].loc[(df["Limiar"]==20)]),axis=1)
limiar_30_ch2 = pd.concat((df['HV'].loc[(df["Limiar"]==30)],df['Cmean'].loc[(df["Limiar"]==30)],df['Cerror'].loc[(df["Limiar"]==30)]),axis=1)
```

```python
df = caract_channel1_inf_plato
```

```python
limiar_10_ch1 = pd.concat((df['HV'].loc[(df["Limiar"]==10)],df['Cmean'].loc[(df["Limiar"]==10)],df['Cerror'].loc[(df["Limiar"]==10)]),axis=1)
limiar_20_ch1 = pd.concat((df['HV'].loc[(df["Limiar"]==20)],df['Cmean'].loc[(df["Limiar"]==20)],df['Cerror'].loc[(df["Limiar"]==20)]),axis=1)
limiar_30_ch1 = pd.concat((df['HV'].loc[(df["Limiar"]==30)],df['Cmean'].loc[(df["Limiar"]==30)],df['Cerror'].loc[(df["Limiar"]==30)]),axis=1)
```

```python
fig2, ax2 = plt.subplots(figsize=(14,8))
ax2.errorbar(x=limiar_10_ch2['HV'], y=limiar_10_ch2['Cmean']/10., yerr=limiar_10_ch2['Cerror']/10., fmt='.', color='r',label='limiar = 10 mV');
ax2.set_title("Platonagem para o cintilador superior")
ax2.set_yscale('log')
ax2.set_xlabel('HV (V)')
ax2.set_ylabel('Counts (Hz)')
ax2.errorbar(x=limiar_20_ch2['HV'], y=limiar_20_ch2['Cmean']/10., yerr=limiar_20_ch2['Cerror']/10., fmt='.', color='g',label='limiar = 20 mV');
ax2.errorbar(x=limiar_30_ch2['HV'], y=limiar_30_ch2['Cmean']/10., yerr=limiar_30_ch2['Cerror']/10., fmt='.', color='b',label='limiar = 30 mV');
ax2.legend()
```

```python
fig1, ax1 = plt.subplots(figsize=(14,8))
ax1.errorbar(x=limiar_10_ch1['HV'], y=limiar_10_ch1['Cmean']/10., yerr=limiar_10_ch1['Cerror']/10., fmt='.', color='r',label='limiar = 10 mV');
ax1.set_title("Platonagem para o cintilador inferior")
ax1.set_yscale('log')
ax1.set_xlabel('HV (V)')
ax1.set_ylabel('Counts (Hz)')
ax1.errorbar(x=limiar_20_ch1['HV'], y=limiar_20_ch1['Cmean']/10., yerr=limiar_20_ch1['Cerror']/10., fmt='.', color='g',label='limiar = 20 mV');
ax1.errorbar(x=limiar_30_ch1['HV'], y=limiar_30_ch1['Cmean']/10., yerr=limiar_30_ch1['Cerror']/10., fmt='.', color='b',label='limiar = 30 mV');
ax1.legend()
```
